### 2) Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
#access maplotlib library and seaborn for the plotting of the graphs
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Get the average of the of the subjects
df['Average'] = df[['Math', 'GEAS', 'Electronics',"Communication"]].mean(axis=1)

#Title of the plot
print('              Features that contributes to average grades')

#create a boxplot with the distribution of data
sns.boxplot(x='Track', y='Average', hue='Gender', data=df)

#render the boxplot
plt.show()
```

```python
#Title for the plot
print('                                 Average Scores by Hometown')

#create the figure along with the dimensions
plt.figure(figsize=(12, 8))

#create a boxplot with the distribution of data
sns.boxplot(x='Hometown', y='Average', data=df)

#render the boxplot
plt.show()
```
